```r
# Création Dataframe
file_data="uppsala_tm_1722-2022.dat"
data_temp <-read.table(file_data)
colnames(data_temp) <- c("Year", "Month", "Day", "Daily_avg", "cor_daily_avg", "id_no")

data_temp$Date <- as.Date(with(data_temp, paste(Year, Month, Day, sep = "-")))
data_temp$Season <- ifelse(data_temp$Month %in% c(12, 1, 2), "Winter",
                    ifelse(data_temp$Month %in% c(3, 4, 5), "Spring",
                    ifelse(data_temp$Month %in% c(6, 7, 8), "Summer", "Autumn")))

data_temp
```

```r
# Valeurs non corrigées
data_metadata <- data.frame(
  mean_tmp = mean(data_temp$Daily_avg, na.rm = TRUE),
  max_tmp  = max(data_temp$Daily_avg, na.rm = TRUE),
  date_max = data_temp$Date[which.max(data_temp$Daily_avg)],
  min_tmp  = min(data_temp$Daily_avg, na.rm = TRUE),
  date_min = data_temp$Date[which.min(data_temp$Daily_avg)],
  nb_NA    = sum(is.na(data_temp$Daily_avg))
)

# Valeurs corrigées pour voir si c'est différent
data_metadata_corr <- data.frame(
  mean_tmp = mean(data_temp$cor_daily_avg, na.rm = TRUE),
  max_tmp  = max(data_temp$cor_daily_avg, na.rm = TRUE),
  date_max = data_temp$Date[which.max(data_temp$cor_daily_avg)],
  min_tmp  = min(data_temp$cor_daily_avg, na.rm = TRUE),
  date_min = data_temp$Date[which.min(data_temp$cor_daily_avg)]
)

data_metadata
data_metadata_corr
```

```r
# Moyenne annuel
annual_avg <- aggregate(Daily_avg ~ Year, data = data_temp, 
                        FUN = function(x) c(mean = mean(x, na.rm = TRUE),
                                            max = max(x, na.rm = TRUE),
                                            min = min(x, na.rm = TRUE),
                                            std = sd(x, na.rm = TRUE)))

annual_avg <- cbind(annual_avg["Year"], annual_avg$Daily_avg)
colnames(annual_avg) <- c("Year", "mean_tmp", "max_tmp", "min_tmp", "std")
cat("Moyennes annuelles :\n")
#annual_avg
                        
seasonal_avg <- aggregate(Daily_avg ~ Year + Season, data = data_temp,
                          FUN = function(x) c(mean = mean(x, na.rm = TRUE),
                                              max = max(x, na.rm = TRUE),
                                              min = min(x, na.rm = TRUE),
                                              std = sd(x, na.rm = TRUE)))

seasonal_avg <- cbind(seasonal_avg[c("Year", "Season")], seasonal_avg$Daily_avg)
colnames(seasonal_avg) <- c("Year", "Season", "mean_tmp", "max_tmp", "min_tmp", "std")
seasonal_avg$roll_mean <- rollmean(seasonal_avg$mean_tmp, k = 30, fill = NA, align = "center")
                          
cat("\nMoyennes saisonnières :\n")
seasonal_avg

# ca serait pas mal de faire avec moyenne glissante plutot
# Moyenne glissante sur 30 j pour avoir les tendances

annual_avg$roll_mean <- rollmean(annual_avg$mean_tmp, k = 30, fill = NA, align = "center")

plot(mean_tmp ~ Year, data = annual_avg, type = "l",
     xlab = "Année", ylab = "Température moyenne annuelle (°C)",
     main = "Évolution de la température annuelle moyenne")

lines(annual_avg$Year, annual_avg$roll_mean, col = "red", lwd = 2)

legend("topleft", legend = c("Moyenne annuelle", "Moyenne glissante (30 ans)"),
       col = c("black", "red"), lty = 1, lwd = 1)

################### SAISONS #########################
saisons <- unique(seasonal_avg$Season)

couleurs <- c("Winter" = "blue",
              "Spring" = "green",
              "Summer" = "red",
              "Autumn" = "orange")

plot(NA, xlim = range(seasonal_avg$Year), ylim = range(seasonal_avg$roll_mean, na.rm = TRUE),
     xlab = "Année", ylab = "Température moyenne (°C)",
     main = "Températures moyennes saisonnières")

for (s in saisons) {
  saison_data <- seasonal_avg[seasonal_avg$Season == s, ]
  lines(saison_data$Year, saison_data$roll_mean, col = couleurs[s], lwd = 2)
}

legend("topleft", legend = saisons, col = couleurs, lty = 1, lwd = 2, bty = "n")
```

```r
plot_saisons <- function(data, var = "mean_tmp", x_l = "Année", y_l = paste(var), title = "Températures moyennes saisonnières") {
  # Permet d'afficher les 4 saisons sur le meme plot
  # Exemple : 
  #   plot_saisons(seasonal_avg, var = "mean_tmp", x_l = "Année", y_l = "Température moyenne (°C)", title = " ")
  saisons <- unique(data$Season)
  couleurs <- c("Winter" = "blue",
                "Spring" = "green",
                "Summer" = "red",
                "Autumn" = "orange")
  
  # Plot vide
  plot(NA, xlim = range(data$Year), ylim = range(data[[var]], na.rm = TRUE),
       xlab = x_l, ylab = y_l,
       main = title)
  
  # Tracer les lignes pour chaque saison
  for (s in saisons) {
    saison_data <- data[data$Season == s, ]
    lines(saison_data$Year, saison_data[[var]], col = couleurs[s], lwd = 2)
  }
  
  # Légende
  legend("topleft", legend = saisons, col = couleurs, lty = 1, lwd = 2, bty = "n")
}

plot_saisons(seasonal_avg, var = "mean_tmp", x_l = "Année", y_l = "Température moyenne (°C)", title = " ")
```

### Variabilité 
On cherche à voir si il y a une augmentation de la variabilité climatique (plus de fluctuations d’une année à l’autre), donc plus d’événements extrêmes ou de transitions rapides (par exemple, vagues de froid ou de chaleur)

```r
# coef de variabilité 
seasonal_avg$cv_tmp <- with(seasonal_avg, std / mean_tmp)
tapply(seasonal_avg$cv_tmp, seasonal_avg$Season, mean, na.rm = TRUE)

# Il y a une valeur de CV abérente j'essaye de voir pourquoi
for (val in seasonal_avg$cv_tmp) {
  if (val > 10) {
    print(val)
  }
}
# -> Moyenne très proche de 0

subset(seasonal_avg, cv_tmp > 10)

seasonal_avg$cv_tmp <- ifelse(abs(seasonal_avg$mean_tmp) < 0.1, NA, seasonal_avg$cv_tmp)

# Moyenne glissante du CV sur 10 ans
seasonal_avg$cv_roll <- ave(seasonal_avg$cv_tmp, seasonal_avg$Season,
                            FUN = function(x) rollmean(x, 10, fill = NA, align = "center"))
                            
tapply(seasonal_avg$cv_roll, seasonal_avg$Season, mean, na.rm = TRUE)
```

```r
saisons <- unique(seasonal_avg$Season)

couleurs <- c("Winter" = "blue",
              "Spring" = "green",
              "Summer" = "red",
              "Autumn" = "orange")

plot(NA, xlim = range(seasonal_avg$Year), ylim = range(seasonal_avg$roll_mean, na.rm = TRUE),
     xlab = "Année", ylab = "Température moyenne (°C)",
     main = "Températures moyennes saisonnières")

for (s in saisons) {
  saison_data <- seasonal_avg[seasonal_avg$Season == s, ]
  lines(saison_data$Year, saison_data$roll_mean, col = couleurs[s], lwd = 2)
}

legend("topleft", legend = saisons, col = couleurs, lty = 1, lwd = 2, bty = "n")

plot_saisons(seasonal_avg, var = "cv_roll", x_l = "Année", y_l = "CV", title = "CV")
```
